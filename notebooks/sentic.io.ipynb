# # Find concept by part of name

```python
def get_concept_by_partname(partname):
    return sentic_df[sentic_df.index.str.contains(partname)].sort_index()
```

# Get all related concepts 

```python
partname = 'screen'
```

```python
concepts = {}
semantic_col_name = ['semantics1', 'semantics2', 'semantics3', 
                     'semantics4', 'semantics5']
df = sentic_df[sentic_df.index.str.contains(partname)].sort_index()
for row in df.iterrows():
    related_concepts = []
    for col in semantic_col_name:
        related_concepts.append(row[1][col])
    concepts[row[0]] = related_concepts
```
