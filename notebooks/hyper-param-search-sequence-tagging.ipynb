```python
import pandas as pd
from pathlib import Path

import pickle
```

```python
from collections import namedtuple
```

```python
HyperParam = namedtuple('HyperParam', 'dataset, batch_size, epochs, dropout, hidden_states, f1')
```

```python
def get_f1(s):
    return float(s.split('\n')[4].split('     ')[3])
```

```python
all_hyper_params = []
for file_path in Path('../results/hyper-param-search/glove.840B.300d/').glob('*'):
    with open(file_path.as_posix(), 'rb') as f:
        try:
            model_info = pickle.load(f)
            all_hyper_params.append(HyperParam(
                Path(model_info['test_file']).stem.replace('_poria-test', ''),
                model_info['batch_size'],
                model_info['epoch'],
                model_info['dropout'],
                model_info['tagger_lstm_dims'],
                get_f1(model_info['eval']),
            ))
        except EOFError as e:
            print(file_path)
```

```python
get_f1(model_info['eval'])
```

```python
df = pd.DataFrame(all_hyper_params)
```

```python
df.sort_values('f1', ascending=False)
```
